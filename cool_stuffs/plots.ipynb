## Custom Problem

### Sphere Function

![sphere](https://wikimedia.org/api/rest_v1/media/math/render/svg/a216c9182cc0eea83b52c120de067286749ee886)
![sphere-plot](https://upload.wikimedia.org/wikipedia/commons/thumb/a/a4/Sphere_function_in_3D.pdf/page1-400px-Sphere_function_in_3D.pdf.jpg)

```python
class SphereProblem(BaseProblem):
    def __init__(self, bounds=None):
        super().__init__(bounds=bounds)

    def evaluate(self, x):
        return sum(x**2)/len(x)
```

```python
problem3 = SphereProblem(bounds=[(-5, 5)]*2)
```

```python
problem3.plot3d()
```

### Schaffer Function

![schaffer](https://wikimedia.org/api/rest_v1/media/math/render/svg/2458c352c0c0524648d8ef713bcea4e80df32fd8)
![schaffer-plot](https://upload.wikimedia.org/wikipedia/commons/thumb/4/46/Schaffer_function_2.pdf/page1-400px-Schaffer_function_2.pdf.jpg)

```python
class SchafferProblem(BaseProblem):
    def __init__(self, bounds=None, default_bounds=(-512, 512)):
        super().__init__(2, bounds, default_bounds)
        
    def evaluate(self, x):
        x1, x2 = x[0], x[1]
        nr = np.sin(x1**2-x2**2)**2-0.5
        dr = (1+0.001*(x1**2+x2**2))**2
        return 0.5+(nr/dr)
```

```python
problem4 = SchafferProblem()
```

```python
problem4.plot3d()
```

## Random Function
$$f(x,y) = -(x^{2}+y^{2})+4$$

```python
class RandomFunction(BaseProblem):
    def __init__(self, bounds=None, default_bounds=(-512, 512)):
        super().__init__(2, bounds, default_bounds)
        
    def evaluate(self, x):
        x1, x2 = x[0], x[1]
        return -(x1**2+x2**2)+4
```

```python
problem5 = RandomFunction()
```

```python
problem5.plot3d()
```
